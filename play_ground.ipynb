```python
from graphviz import Digraph
from anytree.exporter import DictExporter
from data.sast.java.src_parser import java_parser
from util.data_structure import Stack

function_path = '/home/zixian/PycharmProjects/semantic_graph_code_code_clone/test.java'

# Parse the function and retrieve the AST
function = java_parser(function_path)
ast = function.sast
root = ast.root

# Initialize a stack for iterative traversal
stack = Stack()
stack.push(root)

# Initialize the Graphviz Digraph
dot = Digraph(comment='AST of Java Function')
node_counter = 0  # Unique ID for each node in Digraph

# Traverse the tree using the stack
while not stack.is_empty():
    current_node = stack.pop()
    
    # Retrieve node data (type and token)
    _node_data = ast.get_node(current_node).data
    current_node_type = _node_data.node_type.strip('\n').replace(',', ' ')
    current_node_value = _node_data.node_token
    label = f"{current_node_type}:{current_node_value}"
    
    # Add the current node to the Digraph
    dot.node(str(current_node), label)  # Unique node identifier using `current_node`
    
    # Get children of the current node, reverse for correct order, and push to stack
    children = ast.children(current_node)
    children.reverse()  # Ensures left-to-right processing
    
    for child in children:
        stack.push(child.identifier)
        
        # Add an edge from the current node to each child
        dot.edge(str(current_node), str(child.identifier))

# Save or render the graph
dot.render('./ast_graph.png', format='png')  # Change to your desired file path and format

print("es")
```
